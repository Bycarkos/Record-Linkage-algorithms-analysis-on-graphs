## Utils

```python
def read_data(path):
    df = pd.read_csv(path, dtype=str)
    df.fillna("-2", inplace=True)
    return df

def prepare_partition(df, *args, **kwargs):
    """
        args represent the constrains to be used for the partitions
        kwargs means de the value of that constrains to used for the partitions
    """
    #assert np.all([1 if key in args else 0 for key in kwargs.keys()]), "some key is not in args to be constrained"
    
    dA  = copy.copy(df)
    dB  = copy.copy(df)
    
    for cns in args:
        constrained_value = kwargs.get(cns, None)
        if constrained_value is None:continue
        else:
            if type(constrained_value) == list:

                if constrained_value[0].isnumeric():

                    #Create Partitions for the year
                    dA_f = constrained_value[0::2]
                    dB_f = constrained_value[1::2]               

                else:dA_f = dB_f = constrained_value

                dA = dA.loc[(dA[cns].isin(dA_f))]
                dB = dB.loc[(dB[cns].isin(dB_f))] 

            else:

                dA = dA.loc[dA[cns] == constrained_value]
                dB = dB.loc[dB[cns] == constrained_value]
          
    return dA, dB


def get_sampling(df: pd.DataFrame, frac: int = 0.1):
    return df.groupby('gt', group_keys=False).apply(lambda x:x.sample(frac=frac))


def write_pickle(fold:dict,path_file:str):
    with open(path_file, 'wb') as handle:
        pk.dump(fold, handle)
    
def read_pickle(path_file):
    with (open(path_file, "rb")) as openfile:
        return pk.load(openfile)
```

## Indexer for the dataloader

```python
#Indexer Class

class Indexer(object):
    
    def __init__(self, dA: pd.DataFrame, dB: pd.DataFrame, mode: str = "train", blocks:list =[]):

        self._dA = dA
        self._dB = dB
   
        self._mode = mode
        self._blocks = blocks
        if self._mode == "test":
            assert len(self._blocks) != 0
        
    def __call__(self):
        indexer = recordlinkage.Index()
        
        if self._mode == "train":
            indexer.full()
            
        else:
            for block in self._blocks:
                
                indexer.block(block, block)
        
        return indexer.index(self._dA, self._dB)
```

## Class with the Experimental setup

```python
#Comparer Class
class Comparer(ABC):
      
    def __init__(self, pairs):
        self._pairs = pairs
        self._comparer = recordlinkage.Compare()
        
    @abstractmethod
    def compare(self):
        pass
    
    
    def get_gt(self, compared_df):
        reseted = compared_df.loc[compared_df["gt"] == 1].reset_index()
        
        tmp = reseted[["level_0", "level_1", "gt"]]
        lv1 = tmp["level_0"].values
        lv2 = tmp["level_1"].values
        
        return pd.MultiIndex.from_arrays([lv1, lv2], names=['id1', 'id2'])
    
class Experiment_1(Comparer):
    
    def __init__(self, pairs):
        super().__init__(pairs)
        print("Making the comparing to train with [Name, Surname, Second_Surname]")
        
   
    def compare(self, dfA, dfB):
        print("getting GT Done")
        self._comparer.exact('id', 'id', label='gt')
        print("Done")
        self._comparer.string('nom', 'nom', method='jarowinkler', threshold=0.85, label='name')
        self._comparer.string('cognom_1', 'cognom_1', method='jarowinkler', threshold=0.9, label='surname')
        self._comparer.string('cognom_2', 'cognom_2', method='jarowinkler', threshold=0.9, label='second_surname')
        
        return self._comparer.compute(self._pairs, dfA, dfB)
    
    def get_gt(self, compared_df):
        
        return super().get_gt(compared_df)
    
class Experiment_2(Comparer):
    
    def __init__(self, pairs):
        super().__init__(pairs)
        print("Making the comparing to train with [Name, Surname, Second_Surname, Cohort]")

    
    def compare(self, dfA, dfB):
        print("getting GT Done")
        self._comparer.exact('id', 'id', label='gt')
        print("Done")
        self._comparer.string('nom', 'nom', method='jarowinkler', threshold=0.85, label='name')
        self._comparer.string('cognom_1', 'cognom_1', method='jarowinkler', threshold=0.9, label='surname')
        self._comparer.string('cognom_2', 'cognom_2', method='jarowinkler', threshold=0.9, label='second_surname')
        self._comparer.exact("cohort", "cohort", label="Cohort")
        
        return self._comparer.compute(self._pairs, dfA, dfB)
    
    def get_gt(self, compared_df):
        return super().get_gt(compared_df)
    
class Experiment_3(Comparer):
    
    def __init__(self, pairs):
        super().__init__(pairs)
        print("Making the comparing to train with [Name, Surname, Second_Surname, Cohort, Job, Hisclass]")
        
    def compare(self, dfA, dfB):
        print("getting GT Done")
        self._comparer.exact('id', 'id', label='gt')
        self._comparer.string('nom', 'nom', method='jarowinkler', threshold=0.85, label='name')
        self._comparer.string('cognom_1', 'cognom_1', method='jarowinkler', threshold=0.9, label='surname')
        self._comparer.string('cognom_2', 'cognom_2', method='jarowinkler', threshold=0.9, label='second_surname')
        self._comparer.string('ocupacio', 'ocupacio', method='cosine', label='ocupacio')
        self._comparer.exact("cohort", "cohort", label="Cohort")
        #self._comparer.exact("HISCLASS", "HISCLASS",label="hisclass")
        #self._comparer.add(recordlinkage.compare.Variable("HISCAM", "HISCAM",label="hiscam"))
        return self._comparer.compute(self._pairs, dfA, dfB)
        

    def get_gt(self, compared_df):
        return super().get_gt(compared_df)    
```

## Utils for the experimental setup and results

```python

def Prepare_Experiment(df_train: pd.DataFrame, experiment:Type[Comparer], *args, **kwargs):
    """
        args represent the constrains to be used for the partitions
        kwargs means de the value of that constrains to used for the partitions
    """
    #Folds PlaceHolder
    Folds = {}
    
    #Training part
    train_A, train_B = prepare_partition(df_train, *args, **kwargs)
    if kwargs.get("reduce", None) is not None:
        train_A_filtered = (train_A.loc[~train_A["id"].isin(["SFLL", "SCC", "CR"])]).sample(frac=kwargs["reduce"],random_state=1)
        train_B_filtered = (train_B.loc[~train_B["id"].isin(["SFLL", "SCC", "CR"])]).sample(frac=kwargs["reduce"],random_state=1)
    else:
        train_A_filtered = (train_A.loc[~train_A["id"].isin(["SFLL", "SCC", "CR"])])
        train_B_filtered = (train_B.loc[~train_B["id"].isin(["SFLL", "SCC", "CR"])])
    
    if kwargs.get("mode", None) == "test" and kwargs.get("blocks") is not None:
        model_index = Indexer(train_A_filtered, train_B_filtered, kwargs.get("mode", None), kwargs.get("blocks", None))
    
    else:
        model_index = Indexer(train_A_filtered, train_B_filtered)
    pairs = model_index()
    print("The train pairs to compare are {}".format(len(pairs)))

    comparer = experiment(pairs)
    x = time()
    comparation = comparer.compare(train_A_filtered, train_B_filtered)
    print("Comparation done after {}s".format(time()-x))
    sleep(2)
    print("Creating Folds")    
    #Sampling mode to train 
    if kwargs.get("frac", None) is not None:
        for i in tqdm(range(int(1/kwargs.get("frac",None)))):
            
            data_sampling = get_sampling(comparation)
            mdx = comparer.get_gt(data_sampling)
            
            train_X = data_sampling#.loc[:, data_sampling.columns != "gt"]
            
            Folds[i] = (train_X, mdx)
        
        return Folds
    
    if kwargs.get("mode", None) is None or kwargs.get("mode", None) == "train":
        mdx = comparer.get_gt(comparation)
        train_X = comparation#.loc[:, comparation.columns != "gt"]
    
        Folds[0] = (train_X, mdx)
        
    else:
        Folds[0] = (comparation, comparer.get_gt(comparation))
    
    return Folds



def SingRank_Model_Comparing(experiment1: list, experiment2:list, names:list):
    scores = np.array([experiment1, experiment2])
    left, within, right = bt.signrank(scores, rope=0.01, prior_strength=1,verbose=True, names=names)
    

    samples = bt.signrank_MC(scores, nsamples=100000,rope=0.01)
    fig = bt.plot_posterior(samples,names)
    
    plt.show()
    
def Correlated_Model_Comparing(experiment1: list, experiment2:list, names:list):
    rope=0.01
    scores = np.array([experiment1, experiment2])
    
    left, within, right = bt.correlated_ttest(scores, rope=0.01, verbose=True, names=names)

    samples=bt.correlated_ttest_MC(scores, rope=0.01,runs=10)
    #plot posterior
    snb.kdeplot(samples, shade=True) 
    #plot rope region
    plt.axvline(x=-rope,color='orange')
    plt.axvline(x=rope,color='orange')
    #add label
    plt.xlabel(f'{names} on BALL dataset');
    
def Sign_Test_Model_Comparing(experiment1: list, experiment2:list, names:list):
    scores = np.array([experiment1, experiment2])
    left, within, right = bt.signtest(scores, prior_strength=1 ,rope=0.01, verbose=True, names=names)
    samples = bt.signtest_MC(scores, rope=0.01,  nsamples=100000,prior_strength=1)
    fig = bt.plot_posterior(samples,names)
    plt.show()
    
def Tukey_test(scores:list, groups:list):
    df = pd.DataFrame({'score': scores,
                   'group': np.repeat(groups, repeats=10)}) 
    display(df)
    tukey = pairwise_tukeyhsd(endog=df['score'],
                          groups=df['group'],
                          alpha=0.05)
    
    print(tukey)


def anova(*models_results):
    fvalue, pvalue = stats.f_oneway(*models_results)
    print(pvalue)
    if pvalue < 0.05:
        print("Com que el pvalor es menor que 0.05 vol dir que com a minim algun dels models es estadisticament significatiu respecte els altres")
        
    else:
        print("No hi ha evidencies estadístiques de que hi hagi alguna diferencia del fscore entre els diferents models")
def paired_test_variable(samples1: list, samples2:list, names:list, *nvariables):
    
    data = [samples1, samples2]
    fig7, ax7 = plt.subplots()
    ax7.set_title('Fscore distributions')
    bp1 = ax7.boxplot(data, notch=True)
    ax7.legend([bp1["boxes"][0], bp1["boxes"][1]], names, loc='upper right')

    plt.show()
    test,p_value =   stats.ttest_rel(samples1, samples2)
    
    if p_value < 0.05:
        print("Existeix estadísticament una diferrencia significativa entre els dos experiments")
        print(f"Aquesta diferencia significativa entre els dos experiments radica en afegir {nvariables}")
        print(f"El p_value = {p_value}")
        
    else:
        print("no hi ha differencia significativa entre aquests dos experiments")




    
```

## Classical record_linkage models 2 supervised one unsupervised

```python
class Base_Model(ABC):
    def __init__(self, compared_df: pd.DataFrame, gt:pd.MultiIndex, mode: str = "train"):
        
        self._df = compared_df
        self._gt = gt
        self._mode = mode
        
        #evaluating metrics
        self._auc_score = None
        self._confusion_matrix = None
        self._fscore = None
        
    @abstractmethod
    def fit(self):
        pass
    
    def evaluate_model(self,test_df: pd.DataFrame, test_gt: pd.MultiIndex):

        assert 'gt' in test_df, "Not gt column standard name given"
        
        
        #Computing roc auc
        if self.__name__() not in   ["SVM_Model_RL"]:
            
            y_probs = self._model.prob(test_df.loc[:, test_df.columns != 'gt']).values
            y_true  = test_df["gt"].values
            self._auc_score = roc_auc_score(y_true, y_probs)
        
        
        #Computing confusion matrix
        results = self._model.predict(test_df.loc[:, test_df.columns != 'gt'])
        self._confusion_matrix = recordlinkage.confusion_matrix(test_gt, results, len(results))   
        #Fscore
        self._fscore = recordlinkage.fscore(test_gt, results)   
        

        
    def __name__(self):
        raise NotImplemented
        
    #TODO
    def plot_compare(self):
        pass
        
class Logistic_Model_RL(Base_Model):
    
    def __init__(self, compared_df:pd.DataFrame, gt:pd.MultiIndex, mode: str = "train"):
        super().__init__(compared_df, gt, mode)
                
    def fit(self):
        to_train = self._df.loc[:, self._df.columns != 'gt']
        self._model = recordlinkage.LogisticRegressionClassifier()
        self._model.fit(to_train, self._gt)
        
    def evaluate_model(self, test_df: pd.DataFrame, test_gt: pd.MultiIndex):
        super().evaluate_model(test_df, test_gt)
    
    @property
    def coef_(self):
        return self._model.coefficients
    
    @property
    def intercept_(self):
        return self._model.intercept
    
    def __name__(self):
        return "Logistic_Model_RL"

            
            
class SVM_Model_RL(Base_Model):
    
    def __init__(self, compared_df:pd.DataFrame, gt:pd.MultiIndex, mode: str = "train"):
        super().__init__(compared_df, gt, mode)
                
    def fit(self):
        to_train = self._df.loc[:, self._df.columns != 'gt']
        self._model = recordlinkage.SVMClassifier(loss="hinge")
        self._model.fit(to_train, self._gt)
        
    def evaluate_model(self, test_df: pd.DataFrame, test_gt: pd.MultiIndex):
        super().evaluate_model(test_df, test_gt)

        
    def __name__(self):
        return "SVM_Model_RL"

            
            
class ECM_Model_RL(Base_Model):
    
    def __init__(self, compared_df:pd.DataFrame, gt:pd.MultiIndex, mode: str = "train"):
        super().__init__(compared_df, gt, mode)
        self._results = None
                
    def fit(self):
        to_train = self._df.loc[:, self._df.columns != 'gt']
        self._model = recordlinkage.ECMClassifier(init="jaro", max_iter=300,binarize=0.8)
        print(to_train)
        self._model.fit(to_train)
        
    def evaluate_model(self, test_df: pd.DataFrame, test_gt: pd.MultiIndex):
        super().evaluate_model(test_df, test_gt)

        
        
    @property
    def log_match_probs(self):
        """
        Log probability P(x_i=1|Match) as described in the FS framework
        """
        return self._model.log_m_probs

    @property
    def log_p(self):
        """
        Log match probability
        """
        return self._model.log_p
    
    @property
    def log_u_probs(self):
        """
        Log probability P(x_i=1|Non-match)
        """
        return self._model.log_u_probs
    @property
    def log_weights(self):
        return self._model.log_weights
    @property
    def m_probs(self):
        """
        Probability P(x_i=1|Match)
        """
        return self._model.m_probs
    
    @property
    def p(self):
        """
        Match probability
        """
        return self._model.p
    
        
    def __name__(self):
        return "ECM_Model_RL"
    

```

# Defining simple dense model with record_linkage

```python
            
class NN_Model_RL(KerasAdapter, BaseClassifier):
    """Neural network classifier."""
    def __init__(self, input_features:int):

        super(NN_Model_RL, self).__init__()

        
        
        model = tf.keras.Sequential()
        model.add(layers.Dense(16, input_dim=input_features, activation='relu'))
        model.add(layers.Dense(3, activation='relu'))
        model.add(layers.Dense(1, activation='sigmoid'))
        model.compile(
        optimizer=tf.keras.optimizers.Adam(0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
        )
        self.kernel = model
            
            
    def __name__(self):
        return "NN_Model_RL"
    
def evaluate_model_NN(test_df: pd.DataFrame, test_gt: pd.MultiIndex, Model:Type[NN_Model_RL]):
    assert 'gt' in test_df, "Not gt column standard name given"
    auc_score = None
    confusion_matrix = None
    fscore = None


    
    y_probs = Model.prob(test_df.loc[:, test_df.columns != 'gt']).values
    y_true  = test_df["gt"].values
    auc_score = roc_auc_score(y_true, y_probs)


    #Computing confusion matrix
    results = Model.predict(test_df.loc[:, test_df.columns != 'gt'])
    confusion_matrix = recordlinkage.confusion_matrix(test_gt, results, len(results))   
    #Fscore
    try:
        fscore = recordlinkage.fscore(test_gt, results)  
    except:
        fscore = None
    
    return auc_score, confusion_matrix, fscore
```

## Main Function to train the different models

```python
    
def get_metrics(Folds:dict, Folds_test:dict ,Model:Type[Base_Model]):

    auc = []
    conf = []
    fscore = []

    for i in range(len(Folds.keys())):
        train_split, gt_split = Folds[i][0], Folds[i][1]
        
        if Model.__name__ == "NN_Model_RL":
            model = Model(len(Folds[i][0].columns))
            model.fit(train_split, gt_split)
            auc_value, confusion_value, fscore_value = evaluate_model_NN(Folds_test[0][0], Folds_test[0][1], model)
            auc.append(auc_value)
            conf.append(confusion_value)
            fscore.append(fscore_value)
            
        else:
            
            model = Model(train_split, gt_split,"train")
            model.fit()
            
            #change if not ECM
            model.evaluate_model(Folds_test[i][0], Folds_test[i][1])

            auc.append(model._auc_score)
            conf.append(model._confusion_matrix)
            fscore.append(model._fscore)
        
    return auc, conf, fscore
        
    
```
